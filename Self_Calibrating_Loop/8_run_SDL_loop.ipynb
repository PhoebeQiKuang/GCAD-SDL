```python
import os
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

from al_automated.config import ActiveLearningConfig
from al_automated.ground_truth_builder import generate_synthetic_truth
from al_automated.loop_controller import ActiveLearningLoop
from al_automated.sdl_utils import run_gcad_miner, extract_top_circuits, update_gcad_library
```

```python
# ==========================================
# VISUALIZATION HELPER
# ==========================================
def draw_circuit(circuit_dict, cycle_num):
    """Draws simple network graphs of the circuits selected by GCAD."""
    fig, axes = plt.subplots(1, len(circuit_dict), figsize=(4 * len(circuit_dict), 4))
    if len(circuit_dict) == 1: axes = [axes]
    
    TOPO_COLORS = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
    
    for i, (c_name, circuit) in enumerate(circuit_dict.items()):
        ax = axes[i]
        color = TOPO_COLORS[i % len(TOPO_COLORS)]
        
        G = nx.DiGraph(circuit.edge_list)
        pos = nx.spring_layout(G, seed=25)
        
        nx.draw(G, pos, ax=ax, node_color='white', edgecolors=color, linewidths=2, node_size=600, 
                with_labels=True, font_weight='bold', font_size=10, edge_color=color, width=2, arrowsize=15)
        
        ax.set_title(f"{c_name}", color=color, fontweight='bold', fontsize=13)
        
    plt.suptitle(f"SDL Cycle {cycle_num}: Selected Circuits", fontsize=15, y=1.05)
    plt.tight_layout()
    plt.show()

# ==========================================
# THE AUTONOMOUS SDL LOOP
# ==========================================

# 1. AL Configuration
my_custom_dosages = list(np.round(np.arange(0.2, 4.2, 0.4), 2))

config = ActiveLearningConfig(
    max_cycles=3,             
    ensemble_size=50,              
    budget_circuits=2,       
    budget_dosages=2,
    dist_type='lognormal',
    dosages=my_custom_dosages  
)

# 2. Establish the Physical Reality
my_hidden_truth = generate_synthetic_truth(config.nominal_parts_path)

# 3. Directories & Limits
MAX_SDL_CYCLES = 3
gcad_settings_path = os.path.abspath("GCAD/FreeSearch_settings.json")
gcad_engine_dir = os.path.abspath("GCAD")
results_base_dir = os.path.abspath("sdl_results")


for sdl_cycle in range(MAX_SDL_CYCLES):
    print(f"\n\n{'='*40}")
    print(f" 🧬 SDL CYCLE {sdl_cycle}")
    print(f"{'='*40}")
    
    # Step A: GCAD Discovers Architectures
    cycle_out_folder = os.path.join(results_base_dir, f"cycle_{sdl_cycle}")
    run_gcad_miner(gcad_settings_path, cycle_out_folder, gcad_engine_dir)
    
    # Step B: Extractor Selects & Draws Circuits
    my_circuits = extract_top_circuits(cycle_out_folder, m_target=3)
    draw_circuit(my_circuits, sdl_cycle)
    
    # Step C: AL Calibration Engine
    engine = ActiveLearningLoop(my_circuits, my_hidden_truth, config)
    calibrated_parts, _ = engine.run()
    
    # Step D: Update the GCAD Library
    update_gcad_library(calibrated_parts, gcad_engine_dir)

print("\n🎉 ALL SDL CYCLES FINISHED!")
```
