```python
import os
import pickle
import numpy as np

current_dir = os.getcwd()

# Define Paths
nominal_parts_path = os.path.join(current_dir, 'GCAD', 'parts_nominal_backup.pkl')
ground_truth_dir = os.path.join(current_dir, 'ground_truth')

if not os.path.exists(ground_truth_dir):
    os.makedirs(ground_truth_dir)

true_parts_path = os.path.join(ground_truth_dir, 'true_parts.pkl')

print("✅ Paths configured.")
```

```python
# 1. Load the nominal parameters
with open(nominal_parts_path, 'rb') as f:
    nominal_parts = pickle.load(f)

# 2. Create an exact copy to become our "Hidden Truth"
true_parts = nominal_parts.copy()

print("--- DEFINING THE HIDDEN TRUTH ---")

# ==========================================
# MUTATION 1: Altering Z6 (Activator)
# Nominal Z6: [2.0e-02, 5.8e+01, 4.3e-02]
# # Kd gets 25% stronger (0.015), ymax gets 25% stronger (72.5)
# ==========================================
old_z6 = nominal_parts['Z6'].copy()
# Change index 0 and index 1
true_parts['Z6'] = np.array([0.015, 72.5, old_z6[2]])

print(f"🧬 Z6 Mutated:")
print(f"   Nominal: {old_z6}")
print(f"   TRUE:    {true_parts['Z6']}")

# ==========================================
# MUTATION 2: Altering I13 (Inhibitor)
# Nominal I13: [0.012]
# Kd gets 25% stronger (0.009)
# ==========================================
old_i13 = nominal_parts['I13'].copy()
true_parts['I13'] = np.array([0.009])

print(f"\n🧬 I13 Mutated:")
print(f"   Nominal: {old_i13}")
print(f"   TRUE:    {true_parts['I13']}")

print("\n(All other parts remain at nominal values).")
```

```python
with open(true_parts_path, 'wb') as f:
    pickle.dump(true_parts, f)

print(f"✅ Saved 'true_parts.pkl' to {ground_truth_dir}/")
```
